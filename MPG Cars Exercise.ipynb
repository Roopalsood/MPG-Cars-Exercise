```python
import pandas as pd
import numpy as np
```

```python
#The dataset has been imported from quipsamora pandas exercise
cars1 = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv')
cars2 = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv')
cars1.head()
```

```python
#our objective is to remove columns which are unnamed

cars1.drop(columns = ['Unnamed: 9', 'Unnamed: 10',
       'Unnamed: 11', 'Unnamed: 12', 'Unnamed: 13'], inplace=True)

"""
we removed the unnamed columns
making the inplace =true will modify the data for further use 
Using the inplace=True keyword in a pandas method changes the default behaviour such that,
the operation on the dataframe doesn’t return anything, 
it instead ‘modifies the underlying data’ (more on that later). It mutates the actual object which you apply it to.
This means that any other objects referencing this dataframe (such as slices),
will now see the modified version of the data — not the original.
"""

cars1.head()
```

```python
#the number of observation in both the datasets
print(len(cars1), len(cars2))
```

```python
#our next task is to concatinate both cars1, cars2 in one df cars
assert all(cars1.columns == cars2.columns) 
#since it did not show us any flag its true all columns are equal
```

```python
cars = pd.concat([cars1, cars2], axis = 0)
cars.head()

#Axes are defined for arrays with more than one dimension. 
#A 2-dimensional array has two corresponding axes: 
#the first running vertically downwards across rows (axis 0), 
#and the second running horizontally across columns (axis 1)
#Here axis = 0 refers to stacking one over other. 
```

```python
cars.shape #concatenated 
```

```python
#next exercise is to add a column into cars named owner where value is randomly chosen between 15000, 73000.

owners = np.random.randint(15000, 73001, size =398)
len(owners)
```
